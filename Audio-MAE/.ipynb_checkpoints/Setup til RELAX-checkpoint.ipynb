```python
path = r"C:\Users\johau\Desktop\tmp\Masks"
sys.path.insert(1, path)
from Masks import MaskGenerator
from random_walk_mask import Random_Walk_Mask
from segment_mask import Segment_Mask, Segment_Mask2
import torch.nn as nn

#Input data 
wav_file = r"file_example_WAV_2MG.wav"
fbank = wav2fbank(wav_file)
fbank = norm_fbank(fbank)
x = torch.tensor(fbank)
min_x = x.min()

# make it a batch-like (1, 1, 1024, 128)
x = x.unsqueeze(dim=0)
x = x.unsqueeze(dim=0)

MASK_PERCENTAGE = 5
MASKE_TYPE = "Random_walk"
SLIC_seq = ski.segmentation.slic(x[0,0,:,:], n_segments=75, compactness=0.1, sigma = 1, start_label=0, channel_axis=None)

# Similarity function
pdist = nn.CosineSimilarity(dim=1)

# Adjustable parms
mask_bs, num_batches = 1, 1
input_shape = (1024, 128)

# Initialize
model = model_fine
model_name = 'AUDIO-MAE finetuned'
with torch.no_grad():
    h_star = model.forward_encoder_no_mask(x)[:,1:,:].mean(dim=1) # [1, 768]
    
h_star = h_star.expand(mask_bs, -1)

R = torch.zeros(input_shape, dtype=torch.float32) # Importance
U = torch.zeros(input_shape) # Uncertainty

# Loop over batches
for batch_idx in tqdm.trange(num_batches, desc=f"Processing batches for {model_name}"):
    #mask_gen = MaskGenerator(input_shape, mask_type=MASKE_TYPE, mask_percentage=MASK_PERCENTAGE, mask_bs=mask_bs, step_length = 10, max_step_size = 4)
    #mask_gen = Random_Walk_Mask(input_shape, mask_percentage=MASK_PERCENTAGE, mask_bs=mask_bs, step_length = 10, max_change = 4, shutdown_prob = 0.01)
    #mask_gen = Segment_Mask(SLIC_seq, input_shape, mask_percentage=MASK_PERCENTAGE, mask_bs=mask_bs)
    mask_gen = Segment_Mask2(SLIC_seq, input_shape)
    
    batch_masks = mask_gen.generate()
    with torch.no_grad():

        replace_mask = (batch_masks == 0)
        x_mask = x * batch_masks + min_x * replace_mask
        #x_mask = x_mask.squeeze(0)
        #x_mask = x_mask.unsqueeze(1)
        #print(x_mask.shape)

        h = model.forward_encoder_no_mask(x_mask)[:,1:,:].mean(dim=1)   # [mask_bs, 768]
        s = pdist(h_star, h)[:,None, None, None]

        # Update R with the mean over the batch
        #tmp_sum = sum([(s[i] * batch_masks[i]) for i in range(mask_bs)]) / mask_bs
        tmp_sum = (s*batch_masks).mean(dim=0)
    
        R = R.add(tmp_sum)
R /= (num_batches * 0.5)
```

```python
import torch.nn as nn

#Input data 
wav_file = r"file_example_WAV_2MG.wav"
fbank = wav2fbank(wav_file)
fbank = norm_fbank(fbank)
x = torch.tensor(fbank)
min_x = x.min()

# make it a batch-like (1, 1, 1024, 128)
x = x.unsqueeze(dim=0)
x = x.unsqueeze(dim=0)

SLIC_seq = ski.segmentation.slic(x[0,0,:,:], n_segments=75, compactness=0.1, sigma = 1, start_label=0, channel_axis=None)
felz_seq = ski.segmentation.felzenszwalb(x[0,0,:,:], scale = 0.1, min_size = 400)
n_segments = np.max(felz_seq)

# Similarity function
pdist = nn.CosineSimilarity(dim=1)

# Adjustable parms
mask_bs, num_batches = n_segments, 1
input_shape = (1024, 128)

# Initialize
model = model_fine
model_name = 'AUDIO-MAE finetuned'
with torch.no_grad():
    h_star = model.forward_encoder_no_mask(x)[:,1:,:].mean(dim=1) # [1, 768]
    
h_star = h_star.expand(mask_bs, -1)

R = torch.zeros(input_shape, dtype=torch.float32) # Importance
U = torch.zeros(input_shape) # Uncertainty

# Loop over batches
for batch_idx in tqdm.trange(num_batches, desc=f"Processing batches for {model_name}"):
    batch_masks = np.zeros((n_segments, 1, input_shape[0], input_shape[1]))
    for i in range(n_segments):
        batch_masks[i,0,:,:] = (felz_seq == i)
    batch_masks = torch.from_numpy(batch_masks)

    with torch.no_grad():

        replace_mask = (batch_masks == 0)
        x_mask = x * batch_masks + min_x * replace_mask
        x_mask = torch.tensor(x_mask, dtype = torch.float32)

        h = model.forward_encoder_no_mask(x_mask)[:,1:,:].mean(dim=1)   # [10, 768]
        s = pdist(h_star, h)[:,None, None, None]

        # Update R with the mean over the batch
        tmp_sum = sum([(s[i] * batch_masks[i]) / torch.sum(batch_masks[i,0,:,:]) for i in range(n_segments)])
        #tmp_sum = sum([(s[i] * batch_masks[i]) for i in range(n_segments)]) / n_segments
        #tmp_sum = (s*batch_masks).mean(dim=0)
    
        R = R.add(tmp_sum)
#R /= (num_batches * 0.5)
```

```python
plt.imshow(R.T, origin = 'lower')
#plt.colorbar()
print('Minimum pdist value in last batch ', s.min())
print('Index of minimum pdist value: ', s.argmin())
print('Maximum pdist value in last batch ',s.max())
print('Minimum value in importance plot: ', R.min())
print('Maximum value in importance plot: ', R.max())
```
